```python
#Write a program that checks if a number is positive, negative, or zero.

def num_sign(number):
    if number > 0:
        return "Positive"
    elif number<0:
        return "Negative"
    else:
        return "Zero"
    

print((num_sign(+9)))
```

```python
#Create a program that checks if someone is eligible to vote.
def vote_eligibility():
    age = int(input('Enter youre age please'))
    
    if age >= 18:
        print('Can vote')
    else:
        print('Sorry you cant vote')

vote_eligibility()
```

```python
#Write a program that takes 3 numbers and prints the largest one.
def highest_number(num1, num2, num3):
    if num1> num2 and num1>num3:
        print(num1)
    elif num2> num1 and num2> num3:
        print(num2)
    elif num3 > num2 and num3>num1:
        print(num3)

highest_number(6,5,4)
```
