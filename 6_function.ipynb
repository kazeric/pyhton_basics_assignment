```python
#Modify add() to print “even” or “odd” based on the result.
def add(a, b):
    result = a+b
    if result%2 == 0:
        print('even')
    else:
        print('odd')
    return result

add(2,2)
```

```python
# Call a function from within another function.
def multiply(a, b):
    sum = 0
    for i in range(b):
        sum = add(sum, a)
    return sum

print(multiply(4,2))
```
